```python
for sel_session in range(0,len(sessions)):
#     sel_session=0

    block_transitions=np.where(
        sessions[sel_session].trials['trialstimRewarded'][:-1].values!=
        sessions[sel_session].trials['trialstimRewarded'][1:].values)[0]+1
    block_transitions=np.hstack([0,block_transitions,len(sessions[sel_session].trials)-1])

    block_number=np.zeros(len(sessions[sel_session].trials))
    block_number[:]=np.nan

    for bb in range(1,len(block_transitions)):
        block_number[block_transitions[bb-1]:block_transitions[bb]]=bb

    sessions[sel_session].trials['block_number']=block_number

    # trialStimModality=[]
    # rewardedModality=[]
    # # add trialstim modality and rewarded modality
    # for tt in range(0,len(sessions[sel_session].trials)):
    #     trialStimModality.append(sessions[sel_session].trials['trialStimID'].iloc[tt][:-1])
    #     rewardedModality.append(sessions[sel_session].trials['trialstimRewarded'].iloc[tt][:-1])

    # sessions[sel_session].trials['trialStimModality']=trialStimModality
    # sessions[sel_session].trials['rewardedModality']=rewardedModality

    # block_num=[]
    # block_stim_rew=[]
    # block_cross_dprime=[]
    # block_same_dprime=[]

    block_cross_dprime=np.zeros(len(sessions[sel_session].trials))
    block_same_dprime=np.zeros(len(sessions[sel_session].trials))

    # calculate block-by-block dprime
    for bb in np.unique(sessions[sel_session].trials['block_number']):
        if np.isnan(bb):
            continue
        block_trials = sessions[sel_session].trials.query('block_number==@bb and \
                                                          vis_autoreward_trials == False and \
                                                          aud_autoreward_trials == False')

        #base on 1st trial info
        rew_stim=block_trials['trialstimRewarded'].iloc[0]

        if rew_stim=='vis1':

            n_go = len(block_trials.query('trialStimID == "vis1"'))
            n_hit = len(block_trials.query('trialStimID == "vis1" and trial_response'))
            hit_rate=n_hit/n_go

            n_same_nogo = len(block_trials.query('trialStimID == "vis2"'))
            n_same_fa = len(block_trials.query('trialStimID == "vis2" and  trial_response'))
            same_fa_rate=n_same_fa/n_same_nogo

            n_other_nogo = len(block_trials.query('trialStimID != "sound1"'))
            n_other_fa = len(block_trials.query('trialStimID != "sound1" and trial_response'))
            other_fa_rate=n_other_fa/n_other_nogo

        elif rew_stim=='sound1':

            n_go = len(block_trials.query('trialStimID == "sound1"'))
            n_hit = len(block_trials.query('trialStimID == "sound1" and trial_response'))
            hit_rate=n_hit/n_go

            n_same_nogo = len(block_trials.query('trialStimID == "sound2"'))
            n_same_fa = len(block_trials.query('trialStimID == "sound2" and  trial_response'))
            same_fa_rate=n_same_fa/n_same_nogo

            n_other_nogo = len(block_trials.query('trialStimID != "vis1"'))
            n_other_fa = len(block_trials.query('trialStimID != "vis1" and trial_response'))
            other_fa_rate=n_other_fa/n_other_nogo


        if hit_rate==1:
            hit_rate=0.999
        elif hit_rate==0:
            hit_rate=0.001      
        if same_fa_rate==1:
            same_fa_rate=0.999
        elif same_fa_rate==0:
            same_fa_rate=0.001
        if other_fa_rate==1:
            other_fa_rate=0.999
        elif other_fa_rate==0:
            other_fa_rate=0.001

        cross_dprime=(st.norm.ppf(hit_rate) - st.norm.ppf(other_fa_rate))
        same_dprime=(st.norm.ppf(hit_rate) - st.norm.ppf(same_fa_rate))

        block_cross_dprime[block_trials.index]=cross_dprime
        block_same_dprime[block_trials.index]=same_dprime

    #     block_num.append(bb)
    #     block_stim_rew.append(rew_stim)
    #     block_cross_dprime.append(cross_dprime)
    #     block_same_dprime.append(same_dprime)

    # block_info_df=pd.DataFrame({
    #     'block_num':block_num,
    #     'block_stim_rew':block_stim_rew,
    #     'block_cross_dprime':block_cross_dprime,
    #     'block_same_dprime':block_same_dprime,
    # })

    sessions[sel_session].trials['block_cross_dprime']=block_cross_dprime
    sessions[sel_session].trials['block_same_dprime']=block_same_dprime
    #use this to help choose which trials to include

# sessions[sel_session].trials['block_cross_dprime']
```

```python
# distribution of running across different contexts 

bins=np.arange(0,60,2)

run_diffs={
    'mouse':[],
    'session':[],
    'vis_median':[],
    'aud_median':[],
    'p_context':[],
    'first_half_median':[],
    'second_half_median':[],
    'p_halves':[],
}


for sel_session in range(0,len(sessions)):#sessions:
    
    #context
    vis_context_trials=sessions[sel_session].trials.query('trialstimRewarded == "vis1" and \
                                                           vis_autoreward_trials == False and \
                                                           aud_autoreward_trials == False')
    aud_context_trials=sessions[sel_session].trials.query('trialstimRewarded == "sound1" and \
                                                           vis_autoreward_trials == False and \
                                                           aud_autoreward_trials == False')
    vis_context_running=vis_context_trials['avg_run_speed'].values
    aud_context_running=aud_context_trials['avg_run_speed'].values
    h_ctxt,p_ctxt=st.ranksums(vis_context_running, aud_context_running)
    
    #first/second half
    first_half_trials=sessions[sel_session].trials.query('block_number in [0,1,2] and \
                                                           vis_autoreward_trials == False and \
                                                           aud_autoreward_trials == False')
    second_half_trials=sessions[sel_session].trials.query('block_number in [3,4,5] and \
                                                           vis_autoreward_trials == False and \
                                                           aud_autoreward_trials == False')
    first_half_running=first_half_trials['avg_run_speed'].values
    second_half_running=second_half_trials['avg_run_speed'].values
    h_half,p_half=st.ranksums(first_half_running, second_half_running)
    
    run_diffs['mouse'].append(sessions[sel_session].metadata['mouseID'])
    run_diffs['session'].append(sessions[sel_session].metadata['ephys_session_num'])
    run_diffs['vis_median'].append(np.median(vis_context_running))
    run_diffs['aud_median'].append(np.median(aud_context_running))
    run_diffs['p_context'].append(p_ctxt)
    run_diffs['first_half_median'].append(np.median(first_half_running))
    run_diffs['second_half_median'].append(np.median(second_half_running))
    run_diffs['p_halves'].append(p_half)
    
    fig,ax=plt.subplots(2,1,figsize=(6,5),sharex=True)
    #context
    v=ax[0].hist(vis_context_running,bins,alpha=0.5,density=True)
    a=ax[0].hist(aud_context_running,bins,alpha=0.5,density=True)
    
    ctxt_max=np.max(np.hstack([v[0],a[0]]))
    
    ax[0].axvline(np.nanmedian(vis_context_running),color=v[2].get_children()[0].get_facecolor())
    ax[0].axvline(np.nanmedian(aud_context_running),color=a[2].get_children()[0].get_facecolor())
    
    ax[0].set_ylabel('proportion of trials')
    ax[0].legend(['vis context','aud context'],loc='upper right')
    ax[0].text(0,ctxt_max*0.5,'vis median = '+str(np.nanmedian(vis_context_running)))
    ax[0].text(0,ctxt_max*0.4,'aud median = '+str(np.nanmedian(aud_context_running)))
    ax[0].text(0,ctxt_max*0.3,'p='+str(p_ctxt))
    
    #first/second half
    v=ax[1].hist(first_half_running,bins,alpha=0.5,density=True)
    a=ax[1].hist(second_half_running,bins,alpha=0.5,density=True)
    
    half_max=np.max(np.hstack([v[0],a[0]]))
    
    ax[1].axvline(np.nanmedian(first_half_running),color=v[2].get_children()[0].get_facecolor())
    ax[1].axvline(np.nanmedian(second_half_running),color=a[2].get_children()[0].get_facecolor())
    
    ax[1].set_ylabel('proportion of trials')
    ax[1].legend(['first half','second half'],loc='upper right')
    ax[1].text(0,half_max*0.5,'1st median = '+str(np.nanmedian(first_half_running)))
    ax[1].text(0,half_max*0.4,'2nd median = '+str(np.nanmedian(second_half_running)))
    ax[1].text(0,half_max*0.3,'p='+str(p_half))
    ax[1].set_xlabel('run speed')
    
    
    fig.suptitle(sessions[sel_session].metadata['mouseID']+' session '+
                 str(sessions[sel_session].metadata['ephys_session_num']))
    
run_diffs=pd.DataFrame.from_dict(run_diffs)
```
